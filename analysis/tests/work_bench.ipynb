```python
import struct

path = 'binary/09020002.DAT' # Update this to the actual path
data_points = []

with open(path, 'rb') as file: # rb: binary read mode 
    while True:
        data = file.read(2)  # Assuming the data is stored as int16_t (2 bytes)
        if not data:
            break
        value = struct.unpack('<h', data)[0]  # '<' denotes little-endian format, 'h' is int16_t
        data_points.append(value)

print("Data read from file:", data_points)
```

```python
import numpy as np
def t1t2(file_path):
    with open(file_path, 'r') as file:
        # Read the text file
        lines = file.readlines()

        t1 = []
        t2 = []

        # Iterate through lines
        for line in lines[4:]:  # Skip the header
            # Extract timeBefore from the first column (in microseconds)
            time_before = int(line.split(',')[0])
            t1.append(time_before)
            time_after = int(line.split(',')[1])
            t2.append(time_after)

        return t1,t2

t1,t2 = t1t2('M02M4_testing/M0_tLinear/100_500_20/09060001.TXT')

t1 = np.array(t1)
t2 = np.array(t2)

t1 = (t1+t2)/2


# Make a list of the time differences (gaps) between adjacent points:
dt = t1 - np.roll(t1,1)
# skip 1st value
dt = dt[1:]

# t1 = t1 - t1[0]
# print(t1)

tot_len_file = t2[-1]
print(tot_len_file*1e-3)


gap_index_S = np.where(dt > 50)
gap_index_S = gap_index_S[0]
sum_small_gaps = np.sum(dt[gap_index_S])
sum_small_gaps_ms = np.sum(dt[gap_index_S])/1e3
dead_time = (sum_small_gaps/tot_len_file)*100

print(dead_time)
```
